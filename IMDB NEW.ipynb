```python
#import
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
season_no_lst = []
episode_no_lst = []
episode_name_lst = []
episode_airdate_lst = []
episode_score_lst = []
episode_vote_lst = []
episode_desc_lst = []



for season in range(5):
    season_no = season + 1
    
    print(f'PARSING SEASON {season_no}')
    
    #retrieve web data
    url = f'https://www.imdb.com/title/tt2442560/episodes?season={season_no}'
    response = requests.get(url)
    season_page = BeautifulSoup(response.content)
    episode_tiles = season_page.findAll('div', attrs = {'class': 'info'})
    
    #isolate relevant sections
    for episode_no, episode in enumerate(episode_tiles):
        #eps name
        episode_name = episode.strong.a.text
        print(f'episode_name: {episode_name}')

        #episode airdate
        episode_airdate = episode.find('div', attrs ={'class' : 'airdate'}).text.strip().replace('.','')
        print(f'episode_airdate: {episode_airdate} ')

        #episode score
        episode_score = episode.find('div', attrs = {'class': 'ipl-rating-widget'}).find('div',attrs = {'class': 'ipl-rating-star small'}).find('span', attrs = {'class': 'ipl-rating-star__rating'}).text

        print(f'episode_score: {episode_score}')

        #episode votes
        episode_votes = episode.find('div', attrs = {'class': 'ipl-rating-widget'}).find('div',attrs = {'class': 'ipl-rating-star small'}).find('span', attrs = {'class': 'ipl-rating-star__total-votes'
        }).text.strip('()').replace(',', '')

        print(f'episode_votes: {episode_votes}')
        
        #episode desc
        episode_desc = episode.find('div', attrs = {'class' : 'item_description'}).text.strip('\n')
        print(f'episode desc: {episode_desc}')
        print('\n')
        
        season_no_lst.append(season_no)
        episode_no_lst.append(episode_no + 1)
        episode_name_lst.append(episode_name)
        episode_airdate_lst.append(episode_score)
        episode_score_lst.append(episode_score)
        episode_vote_lst.append(episode_votes)
        episode_desc_lst += [episode_desc]
    print('\n')
    

df = pd.DataFrame({
    'season' : season_no_lst,
    'episode_number' : episode_no_lst,
    'title' : episode_name_lst,
    'airdate' : episode_airdate_lst,
    'rating' : episode_score_lst,
    'total_votes' : episode_vote_lst,
    'description' : episode_desc_lst
})
df = df.to_csv('IMDB TABLE.csv')
df
```
